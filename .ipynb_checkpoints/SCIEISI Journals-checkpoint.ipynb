```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')
```

```python
journal = pd.read_csv('SCIEISI Journals.csv')
```

```python
journal.head(2)
```

# Removing Null Category Columns

```python
null_categories = journal.loc[journal['Web of Science Categories'].isnull()]
journal = journal.drop(null_categories.index, axis=0)
```

# Filter Computer Science Journals

```python
journal['Catergories'] = [catergories.replace(' | ', ', ').split(', ') for catergories in journal['Web of Science Categories']]

journal = journal[journal['Catergories'].apply(lambda x: "Computer Science" in x)] 
```

```python
journal
```

```python
journal.to_csv('ISI Journals.csv', index=False)
```
